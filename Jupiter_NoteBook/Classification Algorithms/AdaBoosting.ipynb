# AdaBoosting
   1. AdaBoost is an ensemble learning algorithm that combines many weak learners (usually decision stumps) to create a strong learner.

   2. Key Idea:
Each learner is trained one after the other, and each one tries to fix the mistakes of the previous ones by giving more weight to the misclassified samples.   

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
# Load dataset
data = load_breast_cancer()
dir(data)
```

```python
# Create DataFrame from data.data and add target
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target  # Add target column
```

```python
df.head()
```

```python
# Split Independent Feature X and Dependent Feature y
x = df.drop('target', axis = 'columns')
y = df.target
```

```python
y
```

```python
# Split data into Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Step 4: Create and train AdaBoost model
from sklearn.ensemble import AdaBoostClassifier

model = AdaBoostClassifier(n_estimators=100, learning_rate=0.5, random_state=42)
model.fit(X_train, y_train)
```

```python
# Step 5: Predict and evaluate
from sklearn.metrics import accuracy_score, classification_report
y_pred = model.predict(X_test)
```

```python
print("Accuracy Score:", accuracy_score(y_test, y_pred)*100)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## HypereParameter Tuning using GridSearchCv

```python

from sklearn.model_selection import GridSearchCV
```

```python
# Step 5: Define hyperparameter grid
param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.5, 1.0]
}
```

```python
grid = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
grid = grid.fit(X_train,y_train)
```

```python
# Step 7: Predictions and evaluation
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
# Step 8: Output
print("Best Parameters:", grid.best_params_)
print("Accuracy Score:", accuracy_score(y_test, y_pred)*100)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
