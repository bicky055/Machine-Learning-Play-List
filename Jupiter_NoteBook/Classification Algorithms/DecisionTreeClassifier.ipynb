# Decision Tree in Machine Learning
1. A Decision Tree is a flowchart-like structure used for making decisions or predictions by splitting the data into smaller subsets based on feature values. It’s used in both classification and regression problems.

2. Decision Tree Classifier

       Decision Tree Classifier is a decision tree used for classification tasks (i.e., predicting categories or labels).

       Example: Classify whether a person will buy a car

## Precatical implementation of decision Tree Classifier

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
dataset
```

```python
dataset.target
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# independent feature and dependent feature
x = df.iloc[:,:-1]
y = dataset.target
```

```python
x,y
```

```python
# trian test split 
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33,random_state= 42)
```

```python
x_train 
```

```python
# now train the model using Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

## Post Pruning
1. First grow the full tree, then remove unnecessary branches.

```python
treemodel = DecisionTreeClassifier()
treemodel.fit(x_train,y_train)
```

```python
# In post truning we construct tree fisrt 
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treemodel,filled = True)
plt.show()
```

```python
# here WE see  the 2-depth value[0,34,2] and [0,1,30] so we pruining here

treemodel = DecisionTreeClassifier(max_depth = 2)
treemodel.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (10,6))
tree.plot_tree(treemodel,filled = True)
plt.show()
```

```python
#prediction
y_pred = treemodel.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

## Pre-Pruning 
1. Stops the tree growth early (before it becomes too complex).

```python
# now train the model using Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model
```

```python
## Prepruning
parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter': ['best','random'],
    'max_depth' : [1,2,3,4,5,6],
    'max_features' : ['auto','sqrt','log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(model, param_grid = parameter, cv = 5,scoring = 'accuracy')
cv
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```
