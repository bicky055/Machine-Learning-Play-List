# How to change axis 

```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4]
y = [2,1,3,4]

plt.plot(x,y)
plt.show()
```

# now using xticks() any yticks() fun

```python
x =[1,2,3,4]
y = [2,1,3,4]

plt.bar(x,y)

plt.xticks(x, labels = ["c++","java","C","Python"])
plt.yticks(x)

plt.show()
```

# xlim() and ylim()

```python
x =[1,2,3,4]
y = [2,1,3,4]

plt.bar(x,y)

plt.xlim(0,10)
plt.ylim(0,7)

plt.show()
```

# axis() fun

```python
x =[1,2,3,4]
y = [2,1,3,4]

plt.bar(x,y)

plt.axis([0,8,0,6])

plt.show()
```
