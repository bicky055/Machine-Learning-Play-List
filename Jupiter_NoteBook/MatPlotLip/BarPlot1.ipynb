 # Bar Plot 

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.bar(x,y)
plt.show()
```

```python
x = ["Python",'C++','C','Java']
y = [85,56,68,82]

plt.xlabel("Languages")
plt.ylabel("No")
plt.title("Languages Polularity")
plt.bar(x,y)
plt.show()
```

```python
x = ["Python",'C++','C','Java']
y = [85,56,68,82]

plt.xlabel("Languages",fontsize = 15)
plt.ylabel("No" ,fontsize = 15)
plt.title("Languages Polularity",fontsize = 15)
plt.bar(x,y)
plt.show()
```

```python
# Data
categories = ['Math', 'Science', 'English', 'Hindi', 'Sanskrit']
marks = [85, 90, 78, 56,49]


# add different Colors
c = ['r','b','y','g','m']
# Add labels and title
plt.xlabel('Subjects' , fontsize = 12)
plt.ylabel('Marks', fontsize=12)
plt.title('Marks in Subjects', fontsize = 12)


# Create bar plot
plt.bar(categories, marks, color= c, width = 0.4)
# Show the plot
plt.show()
```

```python
plt.bar(categories, marks, color= c, width = 0.4, align ="edge")  # align = 'center' bydefault
plt.show=()
```

```python
plt.bar(categories, marks, color= c, width = 0.4, edgecolor = 'Black',linewidth = 4 ,linestyle = ":")  
plt.show=()
```

```python
plt.bar(categories, marks, color= c, width = 0.4, edgecolor = 'Black',linewidth = 4 ,linestyle = ":", alpha =0.5)  
plt.show=()
```

```python
plt.barh(categories, marks, color= 'r')

plt.show=()
```

# Plot 2 bar graph in one plot

```python
import matplotlib.pyplot as plt

x = ["Python",'C++','C','Java']
y = [85,56,58,42]
z = [40,30,41,25]

plt.xlabel("Languages", fontsize = 15)
plt.ylabel('No', fontsize = 15)
plt.title("Popularity" , fontsize = 15)
width = 0.2

plt.bar(x,y, width, color = 'r' , label = 'Popularity1')
plt.bar(x,z, width, color = 'y', label = 'Popularity')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python",'C++','C','Java']
y = [85,56,58,42]
z = [40,30,41,25]

p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel("Languages", fontsize = 15)
plt.ylabel('No', fontsize = 15)
plt.title("Popularity" , fontsize = 15)
width = 0.2

plt.bar(p,y, width, color = 'r' , label = 'Popularity1')
plt.bar(p1,z, width, color = 'y', label = 'Popularity')

# show the name of x list 
plt.xticks(p + width/2, x, rotation = 20)
plt.legend()
plt.show()
```
