# Box Plot

```python
import matplotlib.pyplot as plt

data = [10,20,30,40,50,60,70]

plt.boxplot(data)
plt.show()
```

# Dots outliers

```python
data = [10,20,30,40,50,60,70,120]

plt.boxplot(data)
plt.show()
```

 # Horizontal boxPlot

```python
data = [10,20,30,40,50,60,70]

plt.boxplot(data, vert = False)
plt.show()
```

# Shape Of Box

```python
data = [10,20,30,40,50,60,70]

plt.boxplot(data, notch = True)
plt.show()
```

# Labels in list because we can create multiple boxs

```python
data = [10,20,30,40,50,60,70]

plt.boxplot(data, labels =['Pthon'])
plt.show()
```

# fill Box Color 

```python
data = [10,20,30,40,50,60,70]

plt.boxplot(data, labels =['Pthon'], patch_artist=True)
plt.show()
```

# Show means values

```python
data = [10,20,30,40,50,60,70]

plt.boxplot(data,  showmeans = True)
plt.show()
```

```python
data = [10,20,30,40,50,60,70,120]

plt.boxplot(data, showmeans = True,sym="*", flierprops =dict(markeredgecolor = 'gray') )
plt.show()
```

```python
data = [10,20,30,40,50,60,70,120]

plt.boxplot(data, showmeans = True,sym="+", boxprops =dict( color = 'r'),  )
plt.show()
```

```python
data = [10,20,30,40,50,60,70,120]

plt.boxplot(data, showmeans = True,sym="+", capprops =dict(color = 'y') )
plt.show()
```

```python
data = [10,20,30,40,50,60,70,120]

plt.boxplot(data, showmeans = True,sym="+", whiskerprops =dict(color = 'g') )
plt.show()
```

# Multiple Box Plot

```python
data = [10,20,30,40,50,60,70,120]
data2 = [11,22,33,44,55,66]
y =[data, data2]
plt.boxplot(y, labels=['c++','python'] ,showmeans = True,sym="+", capprops =dict(color = 'y') )
plt.show()
```
