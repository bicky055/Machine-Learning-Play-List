# Fill_Between

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.fill_between(x,y)
plt.plot(x,y, color = 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.fill_between(x=[2,4],y1=2,y2=3,color = 'm')
plt.plot(x,y, color = 'k')
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.fill_between(x,y,color = 'm', where = (x>=2) &  (x<=4))
plt.plot(x,y, color = 'k')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Line plot
plt.plot(x, y)

# Fill area between curve and x-axis
plt.fill_between(x, y, color='skyblue', alpha=0.4)

# Labels
plt.title('Fill Between Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show
plt.show()
```
