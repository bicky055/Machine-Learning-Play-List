```python
import matplotlib.pyplot as plt

# Sample data
ages = [22, 25, 26, 27, 30, 32, 35, 36, 40, 42, 45, 48, 50, 52, 60]

# Create histogram
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel("Age Range", fontsize = 15, color = 'purple')
plt.ylabel("Number of People", fontsize = 15, color='purple')
plt.title("Age Distribution",fontsize = 15 , c = 'r')

# Show plot
plt.show()
```

```python
plt.hist(ages, bins=5, color='skyblue', edgecolor='black', histtype = 'step')
plt.show()
```

```python
plt.hist(ages, bins=5, color='skyblue', edgecolor='black', rwidth = 0.8)
plt.show()
```

```python
plt.hist(ages, bins=5, color='skyblue', edgecolor='black', label = 'python')
plt.legend()
plt.axvline(50, c = 'g', label = 'line')
plt.legend()
plt.show()
```
