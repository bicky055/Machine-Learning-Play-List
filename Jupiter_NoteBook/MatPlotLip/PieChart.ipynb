# Pie Chart 

```python
import matplotlib.pyplot as plt

x = [100,500,400,203,142,200]
y = ['Java','Python','C++','C','HTML', 'CSS']
plt.pie(x)
plt.show()
```

# Labels

```python
x = [100,500,400,203,142,200]
y = ['Java','Python','C++','C','HTML', 'CSS']
plt.pie(x, labels = y)
plt.show()
```

# Explode 

```python
x = [100,500,400,203,142,200]
y = ['Java','Python','C++','C','HTML', 'CSS']
ex =[0.3,0.0,0.0,0.0,0.3,0.0]

plt.pie(x, labels = y, explode = ex)
plt.show()
```

# Colors 

```python
x = [100,500,400,203,142,200]
y = ['Java','Python','C++','C','HTML', 'CSS']

c = ['r','g','y','m','b','hotpink','gray']

plt.pie(x, labels = y, colors = c)
plt.show()
```

# Parcentage

```python
x = [100,500,400,203,142,200]
y = ['Java','Python','C++','C','HTML', 'CSS']

c = ['r','g','y','m','b','hotpink','gray']

plt.pie(x, labels = y, colors = c, autopct = '%0.2f%%')
plt.show()
```

# Shadow 

```python
x = [100,500,400,203,142,200]
y = ['Java','Python','C++','C','HTML', 'CSS']

c = ['r','g','y','m','b','hotpink','gray']

plt.pie(x, labels = y, colors = c, autopct = '%0.2f%%', shadow = True,radius = 1.5)
plt.show()
```

# Angle 

```python
x = [100,500,400,203,142,200]
y = ['Java','Python','C++','C','HTML', 'CSS']

c = ['r','g','y','m','b','hotpink','gray']

plt.pie(x, labels = y, colors = c, autopct = '%0.2f%%',labeldistance = 1, startangle = 90 )
plt.show()
```

# labels fontsize and colors

```python
plt.pie(x, labels = y, colors = c, autopct = '%0.2f%%',labeldistance = 1.1, radius = 1.5, 
        textprops ={"fontsize": 15,"color":'m'})
plt.show()
```

```python
plt.pie(x, labels = y, colors = c, autopct = '%0.1f%%',labeldistance = 1.1, radius = 1.5, 
        textprops ={"fontsize": 15,"color":'m'}, counterclock = False, wedgeprops ={'linewidth':5,'width':2})
plt.show()
```

```python
plt.pie(x, labels = y, colors = c, autopct = '%0.1f%%',labeldistance = 1.1, radius = 1.5,shadow = 2, 
        textprops ={"fontsize": 15,"color":'m'}, counterclock = False, wedgeprops ={'linewidth':2,'edgecolor': 'k'})
plt.show()
```

# Center Change

```python
plt.pie(x, labels = y, colors = c, autopct = '%0.1f%%',labeldistance = 1.1, radius = 1.2,shadow = 2, 
        textprops ={"fontsize": 15,"color":'m'}, counterclock = False, wedgeprops ={'linewidth':2,'edgecolor': 'k'},
       center = (3,6))
plt.show()
```

# Rotate labels

```python
plt.pie(x, labels = y, colors = c, autopct = '%0.1f%%',labeldistance = 1.1, radius = 1,shadow = 2, rotatelabels = True )
plt.show()
```

# Titels and legend

```python
plt.pie(x, labels = y, colors = c, autopct = '%0.1f%%',labeldistance = 1.1,shadow = 2)

plt.title("Pie Chart", fontsize = 15)
plt.legend(loc=2, title = 'Language:' )  # loc = 1,2,3,4 only

plt.show()
```

# Doughnut pie charts also call multipal pie chart (ek k andar ek .. etc)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
y1 = np.array([15,24,32,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
c = ['r','b','g','y']

plt.pie(y, labels = mylabels, radius = 1.2, colors=c)
plt.pie(y1, labels = mylabels, radius = 0.5)
plt.show()
```

```python
plt.pie(y, labels = mylabels, radius = 1.2, colors=c)
plt.pie(y1, radius = 0.8,colors='w')
plt.show()
```

# Create a Ring method 2

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
c = ['r','b','g','y']

plt.pie(y, labels = mylabels, radius = 1.5, colors = c)
cr = plt.Circle(xy =(0,0),radius =1, facecolor = 'w')
plt.gca().add_artist(cr)

plt.show()
```
