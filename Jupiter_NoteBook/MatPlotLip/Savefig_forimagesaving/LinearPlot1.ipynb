# Matplotlib Markers used to point edge of the data 

```python
x = [1,2,3,4,5]
y = [ 10,20,35,25,40]

plt.plot(y, marker = 'o')   # Marker = *,^,hexa,..etc
plt.show()
```

```python
plt.plot(y,  'o:r')    # marker|line|color
plt.show() 
```

```python
plt.plot(y,  '*-r')    # marker|line|color
plt.show() 
```

```python
plt.plot(y,  marker = 'o',ms = 20, mec = 'r')  # ms(markerSize), mes(markerEdgeColor)  
plt.show() 
```

```python
plt.plot(y,  marker = 'o',ms = 20, mec = 'r',mfc = 'y')  # ms(markerSize), mes(markerEdgeColor)   mfc(MarkerFaceColor)
plt.show() 
```

# Matplotlib Line 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([23,21,34,54,10])

plt.plot(y, linestyle = 'dotted') 
plt.show()
```

```python
plt.plot(y, ls = '--')  #shoter keys dotted = :, dashed = --, solid = - 
plt.show()
```

```python
plt.plot(y, linestyle = '-', lw = 20)  # lw = linewidth
plt.show() 
```

# Matplotlib Labels and Title

```python
plt.xlabel("Avarge")
plt.ylabel("Number")
plt.title("Student")
plt.plot(y, linestyle = 'dotted') 

plt.show()
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Avarge", fontdict = font1)
plt.ylabel("Number", fontdict = font2)
plt.title("Student" ,fontsize = 20, color = 'r')

plt.plot(y, linestyle = 'dotted') 
plt.show()
```

# Grid 

```python
x = [1,2,3,4,5]
y = [ 10,20,35,25,40]

plt.grid()

plt.plot(x,y, '^--y', ms = 20, mfc = 'hotpink',mec = 'r')  
plt.show()

```

```python
x = [1,2,3,4,5]
y = [ 10,20,35,25,40]

plt.xlabel("Student_no" ,fontsize = 20, color = 'm')
plt.ylabel("Marks")
plt.title("Student Data")

plt.grid()
plt.plot(x,y, '^--y', ms = 20, mfc = 'hotpink',mec = 'black')  

plt.show()
```

```python
x = [1,2,3,4,5]
y = [ 10,20,35,25,40]

plt.xlabel("Student_no" ,fontsize = 20, color = 'm')
plt.ylabel("Marks")
plt.title("Student Data")

plt.grid()
plt.plot(x,y, 'o-y', ms = 20, mfc = 'hotpink',mec = 'black', label = 'students')  
plt.legend()
plt.show()
```

# Matplotlib Subplot Fun

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```
