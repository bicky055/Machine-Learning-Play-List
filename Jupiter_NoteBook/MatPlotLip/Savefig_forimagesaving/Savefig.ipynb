# How to saving files using savefig() fun

```python
import matplotlib.pyplot as plt

x =[2,1,3,4,5,6]
y = [2,1,4,3,5,6]

plt.plot(x,y)

plt.show()
```

# now using savefig() fun 

```python
import matplotlib.pyplot as plt

x =[2,1,3,4,5,6]
y = [2,1,4,3,5,6]

plt.plot(x,y)
plt.savefig('linegraph1')

plt.show()
```

# improve graph quility

```python
import matplotlib.pyplot as plt

x =[2,1,3,4,5,6]
y = [2,1,4,3,5,6]

plt.plot(x,y,color = 'r')

plt.savefig('linegraph2', dpi = 2000)

plt.show()
```

# face color of graph

```python
plt.plot(x,y,color = 'r')

plt.savefig('linegraph3', dpi = 1000,facecolor='y')

plt.show()
```

# Default png file create now create pdf file

```python
plt.plot(x,y,color = 'r')

plt.savefig('linegraph4.pdf', dpi = 1000,facecolor='y')

plt.show()
```

```python
plt.bar(x,y,color = 'm')
plt.savefig('line45.png', dpi=300, bbox_inches='tight', transparent=True, facecolor = 'y')

plt.show()
```
