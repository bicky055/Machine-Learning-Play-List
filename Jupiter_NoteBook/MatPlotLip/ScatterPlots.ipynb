# Creating Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,12,7])
y = np.array([2,4,6,8,10,11,10])

plt.title("Students", fontsize = 20, color = 'm')
plt.xlabel("marks",fontsize = 15, color = 'hotpink')
plt.ylabel("Numbers", fontsize = 15, color = 'y')

plt.scatter(x,y, label = "My_Data")
plt.legend()
plt.show()
```

 # Colors

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,12,7])
y = np.array([2,4,6,8,10,11,10])

colors = ['r','y','g','m','b','r','y']

plt.title("Students", fontsize = 20, color = 'm')
plt.xlabel("marks",fontsize = 15, color = 'hotpink')
plt.ylabel("Numbers", fontsize = 15, color = 'y')

plt.scatter(x,y, label = "My_Data", color = colors)
plt.legend()
plt.show()
```

# Size of Dots

```python

x = np.array([1,2,3,4,5,12,7])
y = np.array([2,4,6,8,10,11,10])

sizes = [100,25,36,150,500,250,990]
colors = ['r','y','g','m','brown','gray','orange']

plt.title("Students", fontsize = 20, color = 'm')
plt.xlabel("marks",fontsize = 15, color = 'hotpink')
plt.ylabel("Numbers", fontsize = 15, color = 'y')

plt.scatter(x,y, label = "My_Data", s=sizes, color=colors)
plt.legend()
plt.show()
```

 # alpha() function use for transparency

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])


plt.scatter(x, y, s=sizes, alpha=0.5, c = colors)

plt.show()
```

  # Markers 

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([9,18,27,36,45,54,63,72])

sizes = np.array([200,500,600,800,400,1000,608,980])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple"])


plt.scatter(x, y, s=sizes, c = colors, marker = '*' ,edgecolor= 'r',linewidth = 2)

plt.show()
```

 # ColorMap

```python
day = [2,3,6,4,1,5]
no = [6,4,8,3,5,7]

sizes =[500,400,800,600,450,1000]
colors =[20,54,23,10,45,80]
plt.xlabel("Days",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title("Scatter Plot",fontsize =15,)


plt.scatter(day,no,s = sizes,cmap = 'viridis', c = colors)
plt.colorbar()
plt.show()
```

 #  Combine Color Size and Alpha

```python
#lets take 100 random values using random.randint() funtion

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

colors = np.random.randint(100, size=(100))
sizes =  8*np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# two scatter in one plot
