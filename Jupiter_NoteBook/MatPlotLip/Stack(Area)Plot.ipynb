# Stack or Area Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,3,1,5,6]
area2 = [5,3,7,4,2]
area3 = [4,2,3,5,1]

plt.stackplot(x,area1,area2,area3, colors = ['r','g','m'])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area1 = [2,3,1,5,6]
area2 = [5,3,7,4,2]
area3 = [4,2,3,5,1]

#labels 
l =['Python','c++','c']
#title
plt.title('Area',fontsize = 15)
plt.xlabel('X-Axis',fontsize=12)
plt.ylabel('Y-axis',fontsize = 12)
plt.stackplot(x,area1,area2,area3, colors = ['r','g','m'], labels = l)
plt.legend(loc=2)
plt.show()
```
