# stem Plot

```python
import matplotlib.pyplot as plt

x = [2,5,3,4,1,6]
y = [2,5,6,1,4,3]

plt.stem(x,y)
plt.show()
```

```python
x = [2,5,3,4,1,6]
y = [2,5,6,1,4,3]

plt.stem(x,y, linefmt = "g:")
plt.show()
```

## Marker 

```python
x = [2,5,3,4,1,6]
y = [2,5,6,1,4,3]

plt.stem(x,y, linefmt = '--', markerfmt = 'r+')
plt.show()
```

## Bottom line

x = [2,5,3,4,1,6]
y = [2,5,6,1,4,3]

plt.stem(x,y, bottom = 1)
plt.show()

## Base line Color Chnage

```python
x = [2,5,3,4,1,6]
y = [2,5,6,1,4,3]

plt.stem(x,y, bottom = 2, basefmt = 'g')
plt.show()
```

## Legend for labels

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 2*np.pi, 10)
y = np.cos(x)

# Create stem plot
plt.stem(x, y, linefmt='--', markerfmt='ro', label = 'Python' )

# Add legend
plt.legend()

plt.title('Stem Plot Example')
plt.show()
```
