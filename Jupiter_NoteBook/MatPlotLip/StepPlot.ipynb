# Step Plot

```python
import matplotlib.pyplot as plt

x =[1,2,3,4,5,6]
y = [2,1,7,3,4,6]

plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x =[1,2,3,4,5,6]
y = [2,1,7,3,4,6]

plt.step(x,y, color='r')
plt.show()
```

# markers

```python
import matplotlib.pyplot as plt

x =[1,2,3,4,5,6]
y = [2,1,7,3,4,6]

plt.step(x,y, color='r', marker ='o',ms = 10,mfc='g')
plt.grid()
plt.show()
```

# labels

```python
import matplotlib.pyplot as plt

x =[1,2,3,4,5,6]
y = [2,1,7,3,4,6]

plt.title('Step Plot')
plt.xlabel('x- axis')
plt.ylabel('y-axis')

plt.step(x,y, color='r',marker = 'o',ms = 12,mfc = 'y',label = 'Python',where = 'mid')
plt.legend(loc = 1)
plt.grid()
plt.show()
```
