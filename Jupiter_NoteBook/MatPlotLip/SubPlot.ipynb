```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)
x_bar = ['A', 'B', 'C', 'D']
y_bar = [10, 20, 15, 5]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Create a big figure
plt.figure(figsize=(12, 8))

# 1. Line Plot
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title("Line Plot")

# 2. Scatter Plot
plt.subplot(2, 3, 2)
plt.scatter(x_scatter, y_scatter, color='r')
plt.title("Scatter Plot")

# 3. Bar Plot
plt.subplot(2, 3, 3)
plt.bar(x_bar, y_bar, color='g')
plt.title("Bar Plot")

# 4. Histogram
plt.subplot(2, 3, 4)
data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple')
plt.title("Histogram")

# 5. Pie Chart
plt.subplot(2, 3, 5)
sizes = [30, 40, 30]
labels = ['Apples', 'Bananas', 'Cherries']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")

# 6. Stem Plot
plt.subplot(2, 3, 6)
markerline, stemlines, baseline = plt.stem([1, 2, 3, 4], [1, 3, 2, 4], linefmt='b-', markerfmt='ro')
plt.title("Stem Plot")

# Adjust layout
plt.tight_layout()

# Show all plots
plt.show()
```
