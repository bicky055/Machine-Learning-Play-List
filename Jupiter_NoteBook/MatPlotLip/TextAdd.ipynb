# How to Add Text on Plots

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [2,1,4,3]

plt.plot(x,y)
plt.title("Languages",fontsize = 15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.show()
```

```python
x = [1,2,3,4]
y = [2,1,4,3]

plt.plot(x,y)
plt.title("Languages",fontsize = 15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,3,'java',fontsize =15,color = 'y',style ='italic',bbox={'facecolor':'r'})

plt.show()
```

```python
x = [1,2,3,4,6]
y = [2,1,4,3,5]

plt.plot(x,y)
plt.title("Languages",fontsize = 15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.ylim(0,7)
plt.annotate("Python", xy =(3,2), xytext =(4,4), arrowprops = dict(facecolor ='k',shrink = 100 ) )

plt.show()
```

```python
x = [1,2,3,4,6]
y = [2,1,4,3,5]

plt.plot(x,y, color= 'r')
plt.title("Languages",fontsize = 15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.ylim(0,7)
plt.legend("up", loc =9, facecolor ='y',edgecolor ='b',framealpha = 0.8 , shadow = True)

plt.show()
```
