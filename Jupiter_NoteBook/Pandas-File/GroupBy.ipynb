# GroupBy Function 

```python
 import pandas as pd
data = pd.DataFrame({ "Name": ['a','b','c','d', 'b','a','a','d','b','c','a','c'],
                     "Subject_1":[12,11,14,15,43,22,11,12,14,36,76,34],
                     "Subject_2":[10,20,32,23,24,54,56,74,34,53,45,14] })
data
```

```python
new_data = data.groupby("Name")
new_data
```

```python
for x,y in new_data:
    print(x)
    print(y)
    print()
```

```python
new_data.get_group('a')
```
