# Interpolate Function

```python
import pandas as pd

data = pd.read_csv("C:\\Users\\VIKRAM\\Desktop\\Jupiter_NoteBook\\leads.csv")
data
```

```python
data.interpolate() # only apply for numeric values
```

```python
data.interpolate(method = 'linear')
```

```python
data.interpolate(method = 'linear', axis = 0)
```

```python
data.interpolate(limit_direction = 'forward',limit = 2) # limit_direction = 'backward' and 'both'  also use
```

```python
data.interpolate(limit_area = 'inside')
```

```python
data.interpolate(limit_area = 'outside') # show all NaN value
```
