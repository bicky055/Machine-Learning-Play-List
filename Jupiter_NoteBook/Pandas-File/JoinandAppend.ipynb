# Join Function  

```python
import pandas as pd

data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]})
data2 = pd.DataFrame({"C":[10,20,30,40], "D":[11,22,33,44]})

data1.join(data2)
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]})
data2 = pd.DataFrame({"C":[10,20], "D":[11,22]})

data1.join(data2)
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]})
data2 = pd.DataFrame({"C":[10,20], "D":[11,22]})

data2.join(data1)
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]} ,index=["a","b","c","d"])
data2 = pd.DataFrame({"C":[10,20], "D":[11,22]}, index = ["c","d"]) # any index give similar as no of elements

data1.join(data2)
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]} ,index=["a","b","c","d"])
data2 = pd.DataFrame({"C":[10,20], "D":[11,22]}, index = ["a","b"])

data2.join(data1, how ='left')
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]} ,index=["a","b","c","d"])
data2 = pd.DataFrame({"C":[10,20], "D":[11,22]}, index = ["a","b"])

data2.join(data1, how ='right', lsuffix = '_12')
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]} ,index=["a","b","c","d"])
data2 = pd.DataFrame({"C":[10,20], "B":[11,22]}, index = ["a","b"])

data2.join(data1, how ='inner',lsuffix = '_12')
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]} ,index=["a","b","c","d"])
data2 = pd.DataFrame({"C":[10,20,30], "B":[11,22,33]}, index = ["a","b","c"])

data2.join(data1, how ='inner',lsuffix = '_12')
```

```python
data1 = pd.DataFrame({"A": [1,2,3,4], "B":[5,6,7,8]} ,index=["a","b","c","d"])
data2 = pd.DataFrame({"C":[10,20,30], "B":[11,22,33]}, index = ["a","b","c"])

data2.join(data1, how ='outer',lsuffix = '_12', rsuffix= '_123')
```

# Melt Fuction for reshaping DataFrame

```python
 import pandas as pd

var = pd.DataFrame({"Name":["bicky","Shivam","Bittu","Ajay","Asheesh"],
                   "Marks":[75,80,82,68,91],
                   "Height":[5.11,5.7,6.0,5.10,6.1]})
var
```

```python
var = pd.DataFrame({"Name":["bicky","Shivam","Bittu","Ajay","Asheesh"],
                   "Marks":[75,80,82,68,91],
                   "Height":[5.11,5.7,6.0,5.10,6.1]})

pd.melt(var)
```

```python
pd.melt(var, id_vars ="Marks")
```

```python
pd.melt(var, id_vars ="Marks",var_name = "Python")
```

```python
pd.melt(var, id_vars ="Marks",var_name = "Python", value_name = "Data")
```

## Pivot fuction used for rearrange rows and columns

```python
data = pd.DataFrame({"Name":["bicky","Shivam","Bittu","Ajay","Asheesh"],
                   "Marks":[75,80,82,68,91],
                   "Height":[5.11,5.7,6.0,5.10,6.1],
                   "Rank":[1,2,3,4,5]})

data
```

```python
data.pivot(index = "Rank", columns = "Name")
```

```python
data = pd.DataFrame({"Name":["a","b","c","a","b","c"],
                   "Marks":[75,80,82,68,91,85],
                   "Height":[5.11,5.7,6.0,5.10,6.1,6.0],
                   "Rank":[1,1,1,1,2,2]})

data.pivot_table(index = 'Name', columns = 'Rank', aggfunc="mean")
```

```python
data.pivot_table(index = 'Name', columns = 'Rank', aggfunc="sum")
```

```python
data.pivot_table(index = 'Name', columns = 'Rank', aggfunc="sum", margins= True)
```
