# Merge Method Used for Merge two DataFrame

```python
import pandas as pd

var1 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
var2 = pd.DataFrame({"A":[1,2,3,4,5], "C":[10,20,30,40,50]})

pd.merge(var1 , var2)
```

```python
pd.merge(var2,var1,on = "A")
```

```python

var3 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
var4 = pd.DataFrame({"A":[1,2,3,4,6], "C":[10,20,30,40,50]})

pd.merge(var3, var4, on = 'A')
```

```python
pd.merge(var3,var4, how = 'left')  # also use right
```

```python
pd.merge(var3,var4, how = 'outer') # inner by default hota hai
```

```python
pd.merge(var3,var4, how = 'outer', indicator = True)
```

```python

var5 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
var6 = pd.DataFrame({"A":[1,2,3,4,6], "B":[10,20,30,40,50]})

pd.merge(var5,var6, left_index = True, right_index = True)
```

```python
pd.merge(var5,var6, left_index = True, right_index = True, suffixes = ("name","id"))
```

# Concat Function Same for Merge Series and DataFrame

```python
sr1 = pd.Series([1,2,3,4,5])
sr2 = pd.Series([11,22,33,44,55])
pd.concat([sr1,sr2])
```

```python
pd.concat([sr1,sr2],axis=1)
```

```python
d1 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3,4,5], "B":[10,20,30,40,50]})
pd.concat([d1,d2])
```

```python
d1 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3], "C":[10,20,30]})
pd.concat([d1,d2], axis = 1)
```

```python
d1 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3], "C":[10,20,30]})
pd.concat([d1,d2], axis = 1, join='inner')
```

```python
d1 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3], "C":[10,20,30]})
pd.concat([d1,d2], axis = 1, join='outer')
```

```python
d1 = pd.DataFrame({"A": [1,2,3,4,5], "B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3,4,5], "B":[10,20,30,40,50]})

pd.concat([d1,d2], axis = 0, keys = ["Frisr","Second"])
```
