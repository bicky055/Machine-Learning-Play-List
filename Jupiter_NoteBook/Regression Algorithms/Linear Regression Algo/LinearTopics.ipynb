# Linear Regression


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([30000, 35000, 40000, 45000, 50000])

# Create Model
model = LinearRegression()
model.fit(x, y)

# Predict
predicted = model.predict(x)

# Plot
plt.scatter(x, y, color='blue')
plt.plot(x, predicted, color='red')
plt.title('Linear Regression')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()

# Print equation
print(f"Equation: Salary = {model.coef_[0]:.2f} * Experience + {model.intercept_:.2f}")
```

# Best fit Line

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([2, 4, 5, 4, 5])

# Create and Train Model
model = LinearRegression()
model.fit(x, y)

# Predict
y_pred = model.predict(x)

# Plot
plt.scatter(x, y, color='blue')    # Actual points
plt.plot(x, y_pred, color='red')    # Best Fit Line
plt.title('Best Fit Line Example')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Print Equation
print(f"Equation: y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}")
```

# Residual Error

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y_actual = np.array([2, 4, 5, 4, 5])

# Train model
model = LinearRegression()
model.fit(x, y_actual)

# Predict
y_pred = model.predict(x)

# Calculate Residuals
residuals = y_actual - y_pred

print("Residuals:", residuals)

# Plot Residuals
plt.scatter(x, residuals, color='red')
plt.axhline(0, color='g', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('x')
plt.ylabel('Residual')
plt.show()
```

# Slope(Coefficient)

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(x, y)

print("Slope (m):", model.coef_[0])
```

# Intercept (C)

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([3, 5, 7, 9, 11])  # Notice: slope 2, intercept 1

model = LinearRegression()
model.fit(x, y)

print("Intercept (c):", model.intercept_)
```
