# DBSCAN (Density-Based Spatial Clustering 

1. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is an unsupervised clustering algorithm that groups together points that are closely packed and marks outliers that lie alone.

2. ðŸ§  Core Ideas:
   1. Clusters = high density areas

   2. Noise = low-density points / outliers


3. How It Works:
   1. For each point:

         1. Count how many points are within eps radius.

         2. If count â‰¥ min_samples, it becomes a core point.

         3. Nearby points are clustered with it.

         4. Points not reachable from any cluster = noise.



# DBSCAN Clustering Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
# Step 1: Create synthetic data
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# Step 2: Standardize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 3: Apply DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.3,min_samples=5)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
# Step 4: Plot the clusters
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan.labels_, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Optional: View cluster labels
print("Cluster labels:", np.unique(dbscan.labels_))
```

# ðŸ” Key DBSCAN Parameters:
1. eps: The max distance between two samples to be considered as neighbors.

2. min_samples: Minimum number of points to form a dense region.

3. dbscan_labels_: -1 means noise/outlier.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=5)  # You can tune these values
dbscan.fit(X_scaled)

# Step 4: Check labels
print("Cluster labels:", np.unique(dbscan.labels_))
print("Assigned labels for each point:", dbscan.labels_)

# Step 5: Visualize clusters using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan.labels_, cmap='plasma')
plt.title("DBSCAN Clustering on Iris (PCA-reduced)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label='Cluster Label')
plt.show()
```

#  Notes:
1. eps=0.6 and min_samples=5 are sensitive â€” feel free to tune them.

2. -1 in dbscan.labels_ means noise points (outliers).

3. We use PCA to visualize the high-dimensional data in 2D.
