# Dimensionality Reduction in Machine Learning
1. Dimensionality Reduction is the process of reducing the number of input features (variables) in a dataset while retaining as much useful information as possible.

2.  Why it's Important:
    1. âœ… Simplifies models â€” makes them faster and easier to understand.

    2. âœ… Reduces overfitting â€” fewer features mean less chance of fitting noise.

    3. âœ… Improves visualization â€” helps us plot high-dimensional data in 2D/3D.

    4. âœ… Speeds up computation â€” fewer features mean faster processing.



## The Curse of Dimensionality in Machine Learning
1. The Curse of Dimensionality refers to problems that arise when your dataset has too many features (dimensions). As dimensions increase, data becomes sparse and less meaningful for machine learning models.


2. âœ… How to Deal with It:
     1. Dimensionality Reduction (e.g., PCA, t-SNE).

     2. Feature Selection (choose only important features).

     3. Regularization (to avoid overfitting).

     4. Use models that handle high dimensions better (e.g., tree-based models).

## PCA (Principal Component Analysis) in ML

1. PCA is an unsupervised machine learning algorithm used for Dimensionality Reduction. It transforms a large set of variables into a smaller one that still contains most of the information.


2. ðŸ§  Intuition Behind PCA:
    1. PCA finds new axes (called principal components) that:

          1. Are linear combinations of the original features.

          2. Capture maximum variance (spread) in the data.

          3. Are uncorrelated (orthogonal) to each other.
      

3. Why Use PCA?
     1. To reduce dimensionality while retaining the most important information.
 
     2. To visualize high-dimensional data (in 2D or 3D).

     3. To remove multicollinearity.

     4. To speed up ML models by reducing input features.



4. How PCA Works (Steps):
     1. Standardize the Data
          1. Make mean = 0 and variance = 1 using StandardScaler.

     2. Compute the Covariance Matrix
          1. Understand how features vary with each other.

     3. Compute Eigenvectors and Eigenvalues

          1. Eigenvectors = directions of principal components

          2. Eigenvalues = amount of variance each component explains

     4. Sort and Select Principal Components
          1. Choose top k components with the highest variance.

     5. Transform Data
          1. Project original data onto the new k components.

# Prectical Implementation for PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
```

```python
df.head()
```

```python
# NOW applying pca Step -1 Standardization of dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
print(scaled_data.shape)
print(scaled_data)
```

```python
# now Step 2 Apping PCA Algo
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)  # 2D
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
# PC(principal components) pc1 > pc2 >pc3 >..... > pcn
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('second principal component')
plt.show()
```
