# UMAP (Uniform Manifold Approximation and Projection)
1. UMAP is a powerful and fast nonlinear dimensionality reduction algorithm, just like t-SNE, but often faster and more scalable.


2. âš™ï¸ How UMAP Works (Simple Intuition):
    1. Constructs a graph of data in high-dimensional space (preserving local relationships).

    2. Projects this graph into a lower-dimensional space.

    3. Tries to preserve the topological structure (local & some global features).

3. ðŸ” Use Cases:
    1. Visualizing embeddings (NLP, images, genes).

    2. Speeding up clustering (e.g., DBSCAN).

    3. Reducing dimensions for supervised learning.

    4. Replacing PCA/t-SNE in preprocessing pipelines.

# Comparing UMAP and t-SNE on the Iris dataset

```python

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import umap
import seaborn as sns

# Load Iris dataset
data = load_iris()
X = data.data
y = data.target

# Apply UMAP
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X)

# Apply t-SNE
tsne_model = TSNE(n_components=2, random_state=42)
X_tsne = tsne_model.fit_transform(X)

# Plotting side-by-side
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# UMAP Plot
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=y, palette="deep", ax=ax[0])
ax[0].set_title("UMAP Projection")

# t-SNE Plot
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, palette="deep", ax=ax[1])
ax[1].set_title("t-SNE Projection")

plt.tight_layout()
plt.show()
```

# 3D Data Visualization in UMAP

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import umap.umap_ as umap
from mpl_toolkits.mplot3d import Axes3D

# Load data
iris = load_iris()
X = iris.data
y = iris.target

# Apply UMAP for 3D
umap_model = umap.UMAP(n_components=3, random_state=42)
X_umap_3d = umap_model.fit_transform(X)

# Plot in 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_umap_3d[:, 0], X_umap_3d[:, 1], X_umap_3d[:, 2], c=y, cmap='viridis')
legend = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend)
ax.set_title("3D UMAP Projection of Iris Dataset")
plt.show()
```
