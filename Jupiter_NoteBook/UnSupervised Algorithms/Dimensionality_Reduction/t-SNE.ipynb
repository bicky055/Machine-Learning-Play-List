# t-SNE (t-distributed Stochastic Neighbor Embedding)
1. -SNE is a powerful nonlinear dimensionality reduction algorithm, mainly used to visualize high-dimensional data in 2 or 3 dimensions.

2. ‚öôÔ∏è How It Works (Simple Intuition):
    1. Compute pairwise similarities between data points in high-dimensional space using probabilities (closer = more similar).

    2. Map points to low-dimensional space so that similar points stay close, and dissimilar points stay apart.

    3. Uses KL Divergence to match similarity distributions between high and low dimensions.

3. üìâ Limitations:
    1. üê¢ Slow for large datasets.

    2. üìè Does not preserve global structure (only local).

    3. ‚ùå Cannot be used for transforming new/unseen data.

    4. üí° Only for visualization, not for modeling.

# Prectical Implementation 

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
X_embedded = tsne.fit_transform(X)
```

```python
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y)
plt.title("t-SNE Visualization of Iris Dataset")
plt.show()
```

## for 3D Visualization

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
# Load Iris dataset
data = load_iris()
X = data.data
y = data.target
```

```python
# Apply 3D t-SNE
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D

tsne_model_3d = TSNE(n_components=3, random_state=42)
X_tsne_3d = tsne_model_3d.fit_transform(X)
```

```python
# Plotting 3D t-SNE result
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_tsne_3d[:, 0], X_tsne_3d[:, 1], X_tsne_3d[:, 2], c=y, cmap='viridis')

legend = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend)
ax.set_title("3D t-SNE Projection")

plt.show()
```
